```python
#1
import numpy as np
from scipy.stats import norm

# Known population parameters
population_mean = 50
population_std = 10

# Sample data
sample = np.random.normal(loc=52, scale=10, size=30)  # Sample mean slightly higher
sample_mean = np.mean(sample)

# Z-test
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(len(sample)))
p_value = 2 * (1 - norm.cdf(abs(z_score)))  # Two-tailed test

# Interpret results
alpha = 0.05
print(f"Z-Score: {z_score:.4f}, P-Value: {p_value:.4f}")
if p_value <= alpha:
    print("Reject the null hypothesis (significant difference)")
else:
    print("Fail to reject the null hypothesis (no significant difference)")
```

```python
#4.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Known population parameters
population_mean = 50
population_std = 10

# Sample data
sample = np.random.normal(loc=52, scale=10, size=30)
sample_mean = np.mean(sample)

# Z-test
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(len(sample)))
p_value = 2 * (1 - norm.cdf(abs(z_score)))

# Visualize decision region
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)
plt.plot(x, y, label="Standard Normal Distribution")
plt.fill_between(x, y, where=(x <= -abs(z_score)) | (x >= abs(z_score)), color="red", alpha=0.5, label="Rejection Region")
plt.axvline(z_score, color="blue", linestyle="--", label="Z-Score")
plt.title("Two-Tailed Z-Test Decision Region")
plt.xlabel("Z-Score")
plt.ylabel("Density")
plt.legend()
plt.show()
```

```python
#5.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
population_mean = 50
population_std = 10
sample_size = 30
alpha = 0.05

# Critical Z-value for two-tailed test
critical_z = norm.ppf(1 - alpha / 2)

# Simulate Type 1 and Type 2 errors
x = np.linspace(40, 60, 1000)
null_dist = norm.pdf(x, loc=population_mean, scale=population_std / np.sqrt(sample_size))
alt_dist = norm.pdf(x, loc=52, scale=population_std / np.sqrt(sample_size))

# Plot
plt.plot(x, null_dist, label="Null Distribution")
plt.plot(x, alt_dist, label="Alternative Distribution")
plt.fill_between(x, null_dist, where=(x <= population_mean - critical_z * (population_std / np.sqrt(sample_size))) | (x >= population_mean + critical_z * (population_std / np.sqrt(sample_size))), color="red", alpha=0.5, label="Type 1 Error")
plt.fill_between(x, alt_dist, where=(x > population_mean - critical_z * (population_std / np.sqrt(sample_size))) & (x < population_mean + critical_z * (population_std / np.sqrt(sample_size))), color="blue", alpha=0.5, label="Type 2 Error")
plt.title("Type 1 and Type 2 Errors")
plt.xlabel("Sample Mean")
plt.ylabel("Density")
plt.legend()
plt.show()
```

```python
#6.
import numpy as np
from scipy.stats import ttest_ind

# Generate two independent samples
sample1 = np.random.normal(loc=50, scale=10, size=30)
sample2 = np.random.normal(loc=55, scale=10, size=30)

# Perform independent T-test
t_stat, p_value = ttest_ind(sample1, sample2)

# Interpret results
alpha = 0.05
print(f"T-Statistic: {t_stat:.4f}, P-Value: {p_value:.4f}")
if p_value <= alpha:
    print("Reject the null hypothesis (significant difference)")
else:
    print("Fail to reject the null hypothesis (no significant difference)")
```

```python
#7.
import numpy as np
from scipy.stats import ttest_rel
import matplotlib.pyplot as plt

# Generate paired samples
before = np.random.normal(loc=50, scale=10, size=30)
after = before + np.random.normal(loc=5, scale=2, size=30)  # Simulate an effect

# Perform paired T-test
t_stat, p_value = ttest_rel(before, after)

# Visualize comparison
plt.scatter(range(len(before)), before, label="Before")
plt.scatter(range(len(after)), after, label="After")
plt.title("Paired Sample Comparison")
plt.xlabel("Observation")
plt.ylabel("Value")
plt.legend()
plt.show()

# Interpret results
alpha = 0.05
print(f"T-Statistic: {t_stat:.4f}, P-Value: {p_value:.4f}")
if p_value <= alpha:
    print("Reject the null hypothesis (significant difference)")
else:
    print("Fail to reject the null hypothesis (no significant difference)")
```

```python
#9.
import numpy as np
from scipy.stats import norm

# Sample data
sample = np.random.normal(loc=50, scale=10, size=30)
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)  # Use ddof=1 for sample standard deviation

# Calculate 95% confidence interval
confidence_level = 0.95
z_critical = norm.ppf((1 + confidence_level) / 2)
margin_of_error = z_critical * (sample_std / np.sqrt(len(sample)))
ci = (sample_mean - margin_of_error, sample_mean + margin_of_error)

print(f"95% Confidence Interval: {ci}")
```

```python
#11
# Bayes' Theorem: P(A|B) = P(B|A) * P(A) / P(B)

# Prior probabilities
P_A = 0.3  # P(A)
P_not_A = 1 - P_A  # P(not A)

# Likelihoods
P_B_given_A = 0.7  # P(B|A)
P_B_given_not_A = 0.2  # P(B|not A)

# Total probability of B
P_B = P_B_given_A * P_A + P_B_given_not_A * P_not_A

# Posterior probability
P_A_given_B = (P_B_given_A * P_A) / P_B

print(f"Posterior Probability P(A|B): {P_A_given_B:.4f}")
```

```python
#19
from scipy.stats import shapiro, levene

# Three random samples
sample1 = np.random.normal(loc=50, scale=10, size=30)
sample2 = np.random.normal(loc=55, scale=10, size=30)
sample3 = np.random.normal(loc=60, scale=10, size=30)

# Normality test (Shapiro-Wilk)
_, p1 = shapiro(sample1)
_, p2 = shapiro(sample2)
_, p3 = shapiro(sample3)

# Homogeneity of variances test (Levene's test)
_, p_levene = levene(sample1, sample2, sample3)

print(f"Normality P-Values: {p1:.4f}, {p2:.4f}, {p3:.4f}")
print(f"Levene's Test P-Value: {p_levene:.4f}")
```

```python
#20
import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
import seaborn as sns

# Simulate data
np.random.seed(42)
data = {
    "value": np.random.normal(loc=50, scale=10, size=60),  # Continuous outcome
    "factor1": np.repeat(["A", "B"], 30),                 # First categorical factor
    "factor2": np.tile(["X", "Y", "Z"], 20)               # Second categorical factor
}

# Create a DataFrame
df = pd.DataFrame(data)

# Perform two-way ANOVA
model = ols("value ~ C(factor1) + C(factor2) + C(factor1):C(factor2)", data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

# Print ANOVA table
print("Two-Way ANOVA Table:")
print(anova_table)

# Visualize interaction between factors
sns.pointplot(x="factor1", y="value", hue="factor2", data=df, ci="sd", capsize=0.1)
plt.title("Interaction Plot: Factor1 vs Factor2")
plt.xlabel("Factor1")
plt.ylabel("Value")
plt.show()
```

```python
#22
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import f_oneway

# Simulate data for three groups
np.random.seed(42)
group1 = np.random.normal(loc=50, scale=10, size=30)  # Group 1
group2 = np.random.normal(loc=55, scale=10, size=30)  # Group 2
group3 = np.random.normal(loc=60, scale=10, size=30)  # Group 3

# Combine data into a DataFrame
data = pd.DataFrame({
    "value": np.concatenate([group1, group2, group3]),
    "group": np.repeat(["Group 1", "Group 2", "Group 3"], 30)
})

# Perform one-way ANOVA
f_stat, p_value = f_oneway(group1, group2, group3)

# Print ANOVA results
print(f"One-Way ANOVA Results:")
print(f"F-Statistic: {f_stat:.4f}")
print(f"P-Value: {p_value:.4f}")

# Interpret results
alpha = 0.05
if p_value <= alpha:
    print("Reject the null hypothesis: There is a significant difference between the groups.")
else:
    print("Fail to reject the null hypothesis: No significant difference between the groups.")

# Visualize the results using boxplots
plt.figure(figsize=(8, 6))
sns.boxplot(x="group", y="value", data=data, palette="Set2")
plt.title("One-Way ANOVA: Comparison of Group Means")
plt.xlabel("Group")
plt.ylabel("Value")
plt.show()
```

```python
#24
from scipy.stats import chi2

# Sample data
sample = np.random.normal(loc=50, scale=10, size=30)
sample_var = np.var(sample, ddof=1)

# Hypothesized population variance
pop_var = 100

# Chi-square test statistic
chi2_stat = (len(sample) - 1) * sample_var / pop_var

# P-value
p_value = 2 * min(chi2.cdf(chi2_stat, len(sample) - 1), 1 - chi2.cdf(chi2_stat, len(sample) - 1))

print(f"Chi-Square Statistic: {chi2_stat:.4f}")
print(f"P-Value: {p_value:.4f}")
```

```python
#26
from scipy.stats import f

# Sample data
sample1 = np.random.normal(loc=50, scale=10, size=30)
sample2 = np.random.normal(loc=55, scale=15, size=30)

# F-test statistic
f_stat = np.var(sample1, ddof=1) / np.var(sample2, ddof=1)

# P-value
p_value = 2 * min(f.cdf(f_stat, len(sample1) - 1, len(sample2) - 1), 1 - f.cdf(f_stat, len(sample1) - 1, len(sample2) - 1))

print(f"F-Statistic: {f_stat:.4f}")
print(f"P-Value: {p_value:.4f}")
```

```python
# 2
import numpy as np
import matplotlib.pyplot as plt

# Generate a discrete uniform distribution
low = 1
high = 6
num_samples = 10000
uniform_samples = np.random.randint(low, high + 1, num_samples)

# Calculate the PMF (Probability Mass Function)
unique_values, counts = np.unique(uniform_samples, return_counts=True)
pmf = counts / num_samples

# Plot the PMF
plt.figure(figsize=(8, 6))
plt.bar(unique_values, pmf)
plt.xlabel("Values")
plt.ylabel("Probability")
plt.title("Probability Mass Function of Discrete Uniform Distribution")
plt.grid(True)
plt.show()
```

```python
#3.
from scipy.stats import bernoulli

def bernoulli_pdf(p, k):
    """
    Calculate the PMF (PDF for discrete) of a Bernoulli distribution.
    :param p: Probability of success
    :param k: Outcome (0 or 1)
    :return: PMF value
    """
    return bernoulli.pmf(k, p)

# Example usage
p = 0.7
print("PMF for k=0:", bernoulli_pdf(p, 0))
print("PMF for k=1:", bernoulli_pdf(p, 1))
```

```python
#4.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters
n, p = 10, 0.5

# Generate random samples
samples = binom.rvs(n, p, size=1000)

# Plot histogram
plt.hist(samples, bins=np.arange(0, n+2) - 0.5, density=True, edgecolor="black")
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.show()
```

```python
#5.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Parameter
lambda_ = 5  # Mean of the Poisson distribution

# Generate random samples
samples = poisson.rvs(lambda_, size=1000)

# Plot histogram
plt.hist(samples, bins=np.arange(0, 20) - 0.5, density=True, edgecolor="black")
plt.title("Poisson Distribution (λ=5)")
plt.xlabel("Number of Events")
plt.ylabel("Probability")
plt.show()
```

```python
#6.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Define the range
low, high = 1, 6

# Calculate CDF
x = np.arange(low, high + 1)
cdf = randint.cdf(x, low, high + 1)

# Plot CDF
plt.step(x, cdf, where="post")
plt.title("CDF of Discrete Uniform Distribution")
plt.xlabel("Outcome")
plt.ylabel("Cumulative Probability")
plt.show()
```

```python
#10
import numpy as np
import matplotlib.pyplot as plt

# Non-normal distribution (e.g., exponential)
population = np.random.exponential(scale=1, size=10000)

# Simulate CLT
sample_means = [np.mean(np.random.choice(population, size=30)) for _ in range(1000)]

# Plot histogram of sample means
plt.hist(sample_means, bins=30, density=True, edgecolor="black")
plt.title("Sampling Distribution of the Mean (CLT)")
plt.xlabel("Sample Mean")
plt.ylabel("Density")
plt.show()
```

```python
#19.
from scipy.stats import norm

# Sample data
sample_mean = 52
population_mean = 50
population_std = 10
sample_size = 30

# Calculate Z-statistic
z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Calculate p-value
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

print(f"Z-Statistic: {z_stat:.4f}, P-Value: {p_value:.4f}")
```

```python
#20.
import numpy as np
from scipy.stats import norm

# Dataset
data = np.random.normal(loc=50, scale=10, size=100)

# Calculate 95% confidence interval
confidence_level = 0.95
mean = np.mean(data)
std_error = np.std(data) / np.sqrt(len(data))
margin_of_error = norm.ppf((1 + confidence_level) / 2) * std_error
ci = (mean - margin_of_error, mean + margin_of_error)

print(f"95% Confidence Interval: {ci}")
```

```python
#25
import numpy as np
import matplotlib.pyplot as plt

# Generate datasets
data1 = np.random.normal(loc=50, scale=10, size=100)
data2 = np.random.normal(loc=50, scale=15, size=100)

# Calculate standard deviations
std1, std2 = np.std(data1), np.std(data2)

# Visualize
plt.hist(data1, bins=30, alpha=0.5, label=f"Dataset 1 (σ={std1:.2f})")
plt.hist(data2, bins=30, alpha=0.5, label=f"Dataset 2 (σ={std2:.2f})")
plt.title("Comparison of Standard Deviations")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
#26
import numpy as np
from scipy.stats import iqr

# Generate dataset
data = np.random.normal(loc=50, scale=10, size=100)

# Calculate range and IQR
data_range = np.max(data) - np.min(data)
data_iqr = iqr(data)

print(f"Range: {data_range:.2f}, IQR: {data_iqr:.2f}")
```

```python
#27
import numpy as np
import matplotlib.pyplot as plt

# Generate dataset
data = np.random.normal(loc=50, scale=10, size=100)

# Z-score normalization
z_scores = (data - np.mean(data)) / np.std(data)

# Visualize
plt.hist(data, bins=30, alpha=0.5, label="Original Data")
plt.hist(z_scores, bins=30, alpha=0.5, label="Z-Score Normalized Data")
plt.title("Z-Score Normalization")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```
